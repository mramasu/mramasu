# Lift: 

Lift gives the correlation between A and B in the rule A=>B. 
Correlation shows how one item-set A effects the item-set B.

Lift(A=>B) = (Support)/(Supp(A)xSupp(B))

If the rule had a lift of 1,then A and B are independent and no rule can be derived from them.

If the lift is > 1, then A and B are dependent on each other, and the degree of which is given by lift value.

If the lift is < 1, then presence of A will have negative effect on B.

In recommendation of video content in streaming platforms, every video has labels or tags of the content. Such as action, horror, adventure in movies. 
Upon the history of the users watching different videos, the current video tags can be related by "Lift Values" with previously watched video tags.

These lift values can be used to predict tags of next recomended video.
